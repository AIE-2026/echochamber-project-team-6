## 6. Primul prompt exploratoriu
Completăm un prompt simplu pentru analizarea comentariilor politice.
Promptul trebuie să ceară:
- ținta comentariului;
- poziționarea față de țintă;
- tonul;
- tema;
- problema de interpretare;
- o justificare scurtă.
Important: tonul sau sentimentul general nu este același lucru cu poziționarea față de țintă.

````python
# IMPORTANT - SCHIMBA PROMTUL DE MAI JOS PENTRU A SE POTRIVI CU CERINȚELE TALE ȘI ASIGURĂ-TE CĂ RESPECTĂ STRUCTURA SOLICITATĂ
# include în prompt instrucțiuni clare pentru fiecare dintre cele 7 elemente pe care vrei să le extragi și asigură-te că modelul înțelege că trebuie să returneze un JSON valid cu exact acele chei
# inlocueste "..." cu instrucțiuni clare pentru fiecare element
# Prompt de sistem: definește rolul modelului
# poti pune si alte axe de analiza care te intereseaza


SYSTEM_PROMPT = """
Ești un expert în științe politice și analiză de text, specializat în detectarea nuanțelor ideologice și a sentimentului în discursul public. 
Rolul tău este să analizezi comentarii politice și să extragi date structurate cu maximă precizie. 
Trebuie să fii obiectiv, să identifici corect entitățile menționate și să detectezi ironia sau sarcasmul care ar putea schimba sensul literal al cuvintelor.
"""

USER_PROMPT_TEMPLATE = """
Citește următorul comentariu politic și identifică:
1. target: Entitatea principală vizată (om politic, partid, instituție sau grup social).
2. stance: Pro fata de target
3. sentiment: Starea emoțională a textului (Pozitiv, Negativ, Furios, Speranzos etc.).
4. tone: Stilul comunicării (Sarcastic, Formal, Agresiv, Umoristic, Disperat).
5. topic: Tema centrală a comentariului (de ex: Economie, Corupție, Alegeri, Sănătate).
6. interpretation_problem: Identifică dacă există ambiguități, jargon specific sau sarcasm care fac interpretarea dificilă (True/False și o scurtă explicație).
7. persuasion_strategy: Tehnica folosită pentru a convinge (de ex: Apel la emoție, Atac la persoană, Date statistice, Generalizare).

Important:
Returnează EXCLUSIV un obiect JSON valid, fără alte explicații introductorii. 
Asigură-te că toate cheile sunt prezente exact așa cum sunt enumerate mai jos.

Returnează JSON valid cu exact aceste chei:"target", "stance", "sentiment", "tone", "topic", "interpretation_problem", "persuasion_strategy"

Comentariu:
<<< {comment_text} >>>
"""
````

## 7. Conectarea la model
Folosim Gemini prin endpoint compatibil cu OpenAI.
Modelul și temperatura au fost setate mai sus.

````python
from openai import OpenAI
client = OpenAI(
    api_key=GEMINI_API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
````

````python
def annotate_comment(comment_text):
    prompt = USER_PROMPT_TEMPLATE.format(comment_text=comment_text)
    response = client.chat.completions.create(
        model=model,
        temperature=temperature,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content
````

## 8. Rulăm promptul pe 10 comentarii
Trimitem fiecare comentariu selectat la model și salvăm răspunsurile.

````python
n_comments = 10  # schimbă aici: 3, 5 sau 10
sample_for_prompt = sample_df.head(n_comments)

outputs = []
for _, row in sample_for_prompt.iterrows():
    outputs.append({
        "source_channel": row.get("source_channel", ""),
        "video_title": row.get("video_title", ""),
        "comment_text": row["text"],
        "model_output": annotate_comment(row["text"])
    })
results_df = pd.DataFrame(outputs)
results_df
````

# 9. Verificam rezultatele

````python
results_df.model_output[0]
````

````python
# funcție pentru a curăța și parsa output-ul modelului, care poate conține JSON în diferite formate (text simplu sau bloc ```json)

def parse_model_output(text):
    # Modelul poate întoarce JSON ca text simplu sau în bloc ```json
    text = text.replace("```json", "")
    text = text.replace("```", "")
    text = text.strip()
    
    return json.loads(text)
````

````python
parsed_outputs = []

for _, row in results_df.iterrows():
    parsed = parse_model_output(row["model_output"])
    
    parsed_outputs.append({
        "source_channel": row["source_channel"],
        "video_title": row["video_title"],
        "comment_text": row["comment_text"],
        "target": parsed.get("target", ""),
        "stance": parsed.get("stance", ""),
        "sentiment": parsed.get("sentiment", ""),
        "tone": parsed.get("tone", ""),
        "topic": parsed.get("topic", ""),
        "interpretation_problem": parsed.get("interpretation_problem", ""),
        "reason": parsed.get("reason", "")
    })

parsed_df = pd.DataFrame(parsed_outputs)
parsed_df
````

### Where did the prompt work well?
A funcționat la mesajele directe și explicite, unde emoția și ținta sunt clar corelate, cum ar fi comentariile de susținere sau atacurile frontale.

### Where did it fail?
A dat greș în identificarea precisă a persoanelor atunci când contextul depindea de titlul videoclipului și nu de textul comentariului în sine.

### Did it confuse sentiment with stance?
Da, a existat tendința de a cataloga poziționarea ca fiind pozitivă doar pentru că autorul folosea un limbaj politicos sau manifesta recunoștință față de jurnaliști, nu față de politician.

### Did sarcasm, ambiguity, or multiple targets create problems?
Sarcasmul a păcălit interpretarea literală a cuvintelor, iar prezența mai multor entități în același text a dus la diluarea verdictului de poziționare.

### What would I change in the next version of the prompt?
Aș cere modelului să identifice separat poziția față de politician versus poziția față de instituția media și să verifice discrepanța dintre cuvintele pozitive și contextul derizoriu.
